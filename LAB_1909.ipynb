**ATIVIDADE LAB 19/09: analisando dados e plotando gráfico**

Objeto de análise: Vendas do álbum "Random Access Memories", de Daft Punk, ao redor do mundo.

fonte: [wikipedia - dados vendas ](https://pt.wikipedia.org/wiki/Random_Access_Memories)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Criando um DICIONÁRIO**

```python
dictVendas = {'Alemanha': 305540,
              'Austrália': 70000,
              'Canadá': 165800,
              'Estados Unidos': 1000000,
              'França': 1200410,
              'Itália': 120588,
              'Japão': 107965,
              'México': 273465,
              'Reino Unido': 309211,
              'Polônia': 41000
              }
```

```python
serie1 = pd.Series(dictVendas)
serie1
```

**Plotando GRÁFICO**

```python
serie1.plot(kind = 'bar')
plt.rcParams["figure.figsize"] = (15,8)
plt.title('Vendas do "Random Access Memories"', fontsize = 25)
plt.xlabel('Países', fontsize = 20)
plt.ylabel('Milhões', fontsize = 20)
plt.legend(['Milhões de Reais'], loc = 'right', fontsize = 15)
plt.show()
```

A partir dessa breve análise, podemos inferir que:


> A França, país de origem do *Duo*, aparece liderando nas vendas do seu último álbum.


> Os Estados Unidos ocuparem a 2ª posição desse ranking
