**Análise 1**

Objeto de análise: Vendas do álbum "Random Access Memories", de Daft Punk, ao redor do mundo.

fonte: [wikipedia - dados vendas ](https://pt.wikipedia.org/wiki/Random_Access_Memories)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Criando um DICIONÁRIO**

```python
dictVendas = {'Alemanha': 305540,
              'Austrália': 70000,
              'Canadá': 165800,
              'Estados Unidos': 1000000,
              'França': 1200410,
              'Itália': 120588,
              'Japão': 107965,
              'México': 273465,
              'Reino Unido': 309211,
              'Polônia': 41000
              }
```

```python
serie1 = pd.Series(dictVendas)
serie1
```

**Plotando GRÁFICO**

```python
serie1.plot(kind = 'bar', color = 'purple')
plt.rcParams["figure.figsize"] = (15,8)
plt.title('Vendas do "Random Access Memories"', fontsize = 25)
plt.xlabel('Países', fontsize = 20)
plt.ylabel('Milhões', fontsize = 20)
plt.legend(['Milhões de Reais'], loc = 'right', fontsize = 15)
plt.show()
```

**A partir dessa breve análise, podemos inferir:**

> O desempenho das vendas de exemplares do álbum em diferentes países;

> A França, país de origem do *Duo*, aparece liderando nas vendas do seu último álbum.

**Análise 2**

Objeto de análise: Gastos dos principais campeonatos de futebol durante a Janela de Transferências 2023.

**fonte:** [Veja Mercado](https://veja.abril.com.br/economia/com-mais-de-r-36-bilhoes-janela-de-transferencias-bate-recorde)

```python
dictGastosTransferencia = {'Premier League': 1980000,
                          'Ligue One': 859000,
                          'Saudi Pro League': 875000,
                          'Bundesliga': 762000,
                          'Calcio': 711000,
                          'La Liga': 405000
                          }
```

```python
serie2 = pd.Series(dictGastosTransferencia)
serie2
```

```python
ax = serie2.plot(kind = 'bar', color = 'red')
for i, v in enumerate(serie2):
    ax.text(i, v + 25000, str(v), ha='center', va='bottom', fontsize=12, color='black')

plt.rcParams["figure.figsize"] = (15,8)
plt.title('Gastos da Janela de Transferências 2023', fontsize = 20)
plt.xlabel('Campeonatos', fontsize = 15)
plt.ylabel('Milhões', fontsize = 15)
plt.legend(['Milhões de Dólares'], loc = 'right', fontsize = 10)
plt.show()
```

**A partir dessa análise, podemos inferir:**


> A Inglaterra segue liderando os gastos nas janelas de transferências há, no mínimo, 8 anos.



> A Saudi Pro League foi o 2º campeonato que mais gastou para a temporada 23/24.


> A liga Árabe está investindo muito para atrair novos e antigos talentos para o seu futebol.
